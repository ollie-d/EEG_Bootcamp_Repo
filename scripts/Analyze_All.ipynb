```python
import numpy as np
import ezbci as ez
import ezbci2 as ez2
import matplotlib.pyplot as plt
import re
```

```python
def MsToFrames(ms, fs):
    dt = 1000 / fs
    return ms / dt
```

```python
# Load data
data_dir = 'C:/Users\H8801/Data/'

fnames = ['OllieRiskitRun000.xdf',
          'OllieRiskitRun001.xdf',
          'OllieRiskitRun002.xdf',
          'OllieRiskitRunSingleSpawn003.xdf',
          'OllieRiskitRunSingleVarWaitSpawn004.xdf', 
          'OllieRiskitRunSingleVarWaitSpawn005.xdf']

for fname in fnames:
    EEG = ez.loadxdf(f'{data_dir}{fname}', False)

    # Estimate photosensor/LSL offset via epoching/averaging
    ERP_stim = ez.epoch(EEG, -100, 500, -100, 500, 'StimPhoto', r'.*_spawn_.*')
    ERP_resp = ez.epoch(EEG, -100, 500, -100, 500, 'RespPhoto', r'R_.*')

    ERP_stim_mean = np.mean(ERP_stim['bin_data'], 2).flatten()
    ERP_resp_mean = np.mean(ERP_resp['bin_data'], 2).flatten()

    stim_delay = np.round(ERP_stim['bin_times'][np.argmax(ERP_stim_mean)] * ERP_stim['fs']/1000).astype(int)
    resp_delay = np.round(ERP_resp['bin_times'][np.argmax(ERP_resp_mean)] * ERP_resp['fs']/1000).astype(int)
    delay = min(stim_delay, resp_delay)

    # Correct time
    EEG['eeg_time'] = EEG['eeg_time'][0:-delay]
    EEG['eeg_data'] = EEG['eeg_data'][delay:]
    
    # Let's analyze
    EEG_filt = ez.filter_eeg(EEG, 0.1, 30, 31, None, 'firwin', causal=False)
    ERN = ez.epoch(EEG_filt, -200, 600, -200, -100, 'Pz', r'.*_invalid')
    CRN = ez.epoch(EEG_filt, -200, 600, -200, -100, 'Pz', r'.*_valid')
    nRN = ez.epoch(EEG_filt, -200, 600, -200, -100, 'Pz', r'.*_corrective')
    
    # Plot
    plt.figure()
    ern_mean = np.mean(ERN['bin_data'], 2).flatten()
    crn_mean = np.mean(CRN['bin_data'], 2).flatten()
    nrn_mean = np.mean(nRN['bin_data'], 2).flatten()
    plt.plot(ERN['bin_times'], ern_mean, 'r-', label=f"Invalid (n={ERN['bin_data'].shape[-1]})")
    plt.plot(CRN['bin_times'], crn_mean, 'b-', label=f"Valid (n={CRN['bin_data'].shape[-1]})")
    plt.plot(nRN['bin_times'], nrn_mean, 'g-', label=f"Corrective (n={nRN['bin_data'].shape[-1]})")
    plt.xlabel('Time (ms)')
    plt.ylabel('Amplitude ($\mu$V) at Pz')
    plt.title('Response ERPs')
    plt.legend(loc='best')
```
